### 대응표본T검정 - 단일모집단에 두개의 처리(전/후)

- 같은 집단이므로 등분산성 만족!
- 모집단의 정규분포 가정(shapiro)

### 귀무가설/대립가설

- H0 : 두 모평균 사이의 차이는 없다.(동일)
- H1 : 두 모평균 사이의 차이는 있다.(유의미한 차이가 있다)

### 검정통계량

```python
import numpy as np
import scipy.stats as stats

d=25 #표본 차이의 평균
s=9.5 #차이의 표본표준편차
n=20 

SE=s/np.sqrt(n) #표준오차(=표본표준편차)
t=d/SE
t
```

->이러한 방식으로 계산됨

### 채택

```python
#t>=0
rv=stats.t(df=n-1)
p=(1-rv.cdf(t))*2
p
```

```python
#임곗값
alpha=0.05
rv=stats.t(df=n-1)
rv.interval(1-alpha)
```

귀무가설을 기각한다. 즉 두 모평균 사이에는 유의미한 차이가 있다.
