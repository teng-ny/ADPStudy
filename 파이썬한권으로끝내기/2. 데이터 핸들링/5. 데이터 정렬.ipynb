```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
iris=pd.DataFrame(iris.data, columns=iris.feature_names)
iris.sort_index(ascending=False, inplace=True)
iris.head()
```

```python
#axis=1:컬럼명 기준, axis=0:행기준
#ascending=True:오름차순, ascending=False:내림차순(큰숫자)
iris.sort_index(axis=1, ascending=True, inplace=True)
iris.head()
```

```python
iris.sort_values('petal length (cm)', ascending=False)
iris
```

```python
iris.sort_values(['petal width (cm)', 'sepal length (cm)'])
```
