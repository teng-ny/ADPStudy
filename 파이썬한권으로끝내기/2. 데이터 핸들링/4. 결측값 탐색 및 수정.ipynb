```python
import pandas as pd
students=pd.DataFrame({'이름':['장화', '홍련', '콩쥐', '팥쥐', '해님', '달님'], '국어':[70, 85,None, 100, None, 85], '수학':[65, 100, 80, 95, None, 70]})
students
```

```python
students[students['이름']=='장화']
```

```python
students[(students['국어']>=80)&(students['수학']>=80)]
```

```python
students.loc[6, '이름':'수학']=['별님',50, 60]
```

```python
students.loc[(students['국어']>=80)&(students['수학']>=70), '합격']='Pass'
#조건을 충족하지 못하는 행은 NaN으로 표시
students.loc[students['합격']!='Pass', '합격']='Fail'
```

```python
students
```

### 한번에 여러개 조건을 반환해야하는 경우-> np.select

```python
import numpy as np
```

```python
condition_list=[(students['국어']>=90), (students['국어']>=80)&(students['국어']<90), (students['국어']>=70)&(students['국어']<80)]
choice_list=['A','B','C']
```

```python
students['점수']=np.select(condition_list, choice_list, default='F')
students
```

### 결측값 대체

```python
health=pd.DataFrame({'연도':[2017,2018,2019,2020,2021,2022], '키':[160,162,165,None, None, 166], '몸무게':[53,52,None,50,51,54], \
                     '시력':[1.2,None, 1.2,1.2, 1.1, 0.8], '병결':[None, None,None,2,None,1]})
health.fillna(0)
```

```python
health.fillna(health.mean())
```

```python
health
```

```python
health['병결']=health['병결'].fillna(0)
health['몸무게']=health['몸무게'].fillna(health['몸무게'].mean())
health
```

```python
#직전값으로 대체하기
health.fillna(method='pad', inplace=True)
health
```

```python
#중복행 삭제
health['키'].drop_duplicates()
```

```python
health[['시력','병결']].drop_duplicates()
```
