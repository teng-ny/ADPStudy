## 그룹화와 집계(group by)

```python
#데이터 불러오기
import pandas as pd
from sklearn.datasets import load_iris
IRIS=load_iris()
iris=pd.DataFrame(data=IRIS.data, columns=IRIS.feature_names)
iris['class']=IRIS.target
iris['class']=iris['class'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
iris
```

```python
iris.groupby(['class']).mean()
```

```python
iris.groupby(['class']).median()
```

## 도수분포표(value_counts())

```python
pd.Series(iris['class']).value_counts()
```

```python
iris['class'].value_counts()
```

```python
iris['petal width level']=pd.qcut(iris['petal width (cm)'], q=3, labels=['short', 'middle','long'])
iris
```

```python
iris['petal width level'].value_counts()
```

```python
#교차표
pd.crosstab(iris['petal width (cm)'], iris['petal width level'])
```
