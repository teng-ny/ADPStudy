## 피벗 테이블(pivot_table( ))

```python
import pandas as pd
score=pd.DataFrame({'학년':[1,1,1,1,2,2], '반':['A', 'A', 'B', 'B','C','C'], '성별':['여자', '남자', '여자','남자','여자',\
                                                                   '남자'], '성적':[76,88,85, 72, 68, 70]})
score
```

```python
score.pivot_table(index=['학년','반'], columns='성별', values='성적')
```

```python
score.melt(id_vars=['학년', '반'], var_name='성별', value_name='성적')
```
