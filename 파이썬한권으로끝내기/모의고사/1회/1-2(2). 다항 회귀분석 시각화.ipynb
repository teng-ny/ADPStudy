단순선형회귀를 다항 회귀로 3차까지 적용시켜 계수를 구하고 3차항을 적용한 모델의 스캐터 플롯과 기울기 선을 그리시오.

```python
import pandas as pd
import numpy as np
```

```python
#수식
m=100
x=6*np.random.rand(m,1)-3
y=3*x**3+x**2+2*x+2+np.random.randn(m,1)
#단순선형회귀(?)
```

```python
line=np.linspace(-3,3,100, endpoint=False).reshape(-1,1)
#endpoint=True:-3,3으로 정확하게 끝나도록
#endpoint=False: 정확하게 끝나지 않더라도 괜찮음
#reshape:재배열
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
poly=PolynomialFeatures(degree=3, include_bias=False)
x_poly=poly.fit_transform(x)
#x는 교차항과 3차항을 포함함

line_poly=poly.transform(line)
#예측할 변수들도 변환

linear=LinearRegression()
reg=linear.fit(x_poly, y)
```

```python
plt.plot(line, reg.predict(line_poly), c='r', linewidth=3)
#기울기선
plt.plot(x,y,'o',c='g',alpha=0.5)
#3차항을 적용한 모델의 스캐터플롯
```

## 다시해보기

```python
import pandas as pd
import numpy as np

m=100
x=6*np.random.rand(m,1)-3
y=3*x**3+x**2+2*x+2+np.random.randn(m,1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(x,y,s=5)
```

다음과 같은 x,y데이터는 분포가 곡선으로 나타나기 때문에\
일반적으로 선형회귀로 해결할 수 없다.\
-> 다항회귀\
: 형태가 비선형인 데이터가 있을 때 데이터에 각 특성의 제곱값을 추가해주어\
특성이 추가된 데이터를 선형회귀 모델로 훈련함.

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3, include_bias=False)
x_poly=poly.fit_transform(x)
#데이터를 다항형태로 변환하여 저장

from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(x_poly, y)
#변형된 데이터와 기존 y값을 넣고 회귀 모델 학습
```

```python
line=np.linspace(-3, 3, 100, endpoint=False).reshape(-1, 1)
line_poly=poly.transform(line)

plt.plot(line, reg.predict(line_poly), c='r', linewidth=3)
plt.plot(x,y,'o',c='g',alpha=0.5)
```
