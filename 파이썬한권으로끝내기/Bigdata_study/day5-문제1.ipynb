- 회사 제품의 금속 재질 함유량의 분산이 1.3을 넘으면 불량이라고 판단한다. 회사에서는 품질경영팀으로부터 제조사별로 금속 함량이 차이가 난다고 제보를 받았다. 해당 금속 함유량 데이터에 대한 검정을 수행하시오.(유의확률 : 0.05)

- 데이터 불러오기

```python
import pandas as pd
import numpy as np
data=pd.read_csv("metalicity")
```

```python
print(round(data['metalicity'].mean(),2))
#평균 45.27
print(len(data))
```

(1) 귀무가설과 대립가설을 작성하시오.

H0 : 분산은 1.3이다.
H1 : 분산은 1.3이 아니다.

(2) 가설을 양측 검정하시오.

- 정규성 검정

가설을 검정하기 전에 해당 데이터가 정규성을 띄는지 먼저 판단해주어야 함.

```python
import scipy.stats as stats
from scipy.stats import shapiro
shapiro(data['metalicity'])
```

p-value가 0.05보다 크므로 귀무가설을 기각하지 못한다. 즉, 정규성을 따른다고 판단할 수 있다.
표본데이터가 정규성을 가지므로 정규분포를 따르는 분산이 1.3인 비교데이터를 만들어주고 등분산 검정을 진행할 수 있다.
* 등분산 검정을 하려면 비교대상이 있어야하기 때문에 비교데이터를 만들어준 후 등분산 검정

```python
#np.random.normal(해당데이터 평균, 표준편차, 행개수)
test=np.random.normal(45.27, 1.3**0.5, 150)
test
```

```python
stats.levene(data['metalicity'], test)
```

등분산 검정 결과 p-value값이 0.05보다 작으므로 귀무가설을 기각한다. 즉, 제품의 금속재질 함유량 분산은 1.3이 아니라고 할 수 있다.
