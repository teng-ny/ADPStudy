```python
import pandas as pd
import numpy as np
```

- 데이터 불러오기

```python
us=pd.read_csv("problem2_usage.csv")
us.head(3)

#timestamp->날짜로 변환
us['time']=pd.to_datetime(us['timestamp'], unit='s')

#날짜 순서대로 정렬
us=us.sort_values('time').reset_index(drop=True)
us.head(3)
```

```python
ush=pd.read_csv('problem2_usage_history.csv')
ush.head(3)
```

```python
ush=ush[ush['hh:mm'].isin(['05:59', '11:59', '17:59', '23:59'])]
ush.loc[:, 'total']=ush.iloc[:, 2:].sum(axis=1).round(4)
```

```python
ush
```

- 조인

```python
ush=ush.rename(columns={'t':'usage'}).reset_index(drop=True)
pd.merge(ush, us)
us
```
