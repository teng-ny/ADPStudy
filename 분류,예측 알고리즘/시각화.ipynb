- 히스토그램

```python
import pandas as pd
df=pd.read_csv("titanic_train.csv")
df.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,3))

sns.histplot(df['Age'], bins=20)
plt.show()
```

- 카운트 플롯(이산형)

```python
sns.countplot(x='Pclass', data=df)
plt.show()
```

```python
sns.barplot(x='Pclass', y='Age', data=df)
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df, ci=None, estimator=sum)
plt.show()
```

```python
#hue파라미터로 Sex를 설정
sns.barplot(x='Pclass', y='Age', hue='Sex', data=df)
plt.show()
```

- 박스 플롯

```python
sns.boxplot(y='Age', data=df)
plt.show()
```

- 바이올린 플롯(히스토그램+박스플롯)

```python
sns.violinplot(y='Age', data=df)
plt.show()
```

- 산점도(scatter plot)

: x, y축 모두 연속형 숫자값을 적용해야 의미있는 시각화 정보를 얻을 수 있음

```python
sns.scatterplot(x='Age', y='Fare', data=df)
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.show()
```

- 히트맵

```python
corr_df=df.corr()
corr_df
```

```python
corr=df.corr()
sns.heatmap(corr)
plt.show()
```

```python
sns.heatmap(corr, annot=True, fmt='.1f', cbar=False)
plt.show()
```

```python
sns.heatmap(corr, annot=True, fmt='.1f', cbar=True)
plt.show()
```
