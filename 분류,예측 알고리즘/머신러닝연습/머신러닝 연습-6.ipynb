```python
df.drop(columns=['WindDir9am', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm'], inplace=True)
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(columns=['RainTomorrow'])
y=df['RainTomorrow']
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=6789)
```

```python
from sklearn.preprocessing import LabelEncoder

features = ['Location','WindGustDir','WindDir3pm', 'RainToday']

for feature in features:
    le = LabelEncoder()
    X_train[feature] = le.fit_transform(X_train[feature])
    X_test[feature] = le.transform(X_test[feature])
    
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```
